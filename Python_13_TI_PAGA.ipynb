```python
### Import Libraries.

import scanpy as sc
import anndata as ad
import os
```

```python
### Load Data.

os.chdir("/folder/")
adata = ad.read_h5ad('adata.h5ad')
```

```python
### Force-directed layout (Draw-graph) Visualization

sc.tl.draw_graph(adata, init_pos = 'X_umap')
sc.settings.set_figure_params(dpi = 800, figsize = (10, 10))
sc.pl.draw_graph(adata, color = 'Cluster_Column', legend_loc = 'on data', legend_fontsize = 'xx-small')
```

```python
### PAGA Trajectory Inference.

sc.tl.paga(adata, groups = 'Cluster_Column')
sc.settings.set_figure_params(dpi = 200, figsize = (10, 10))
```

```python
### Visualize PAGA Connectivity and Hierarchical Structure.

sc.pl.paga(
    adata,
    color = 'Cluster_Column',
    layout = "rt",
    root = [Number_of_Root],
    node_size_scale = 30,
    edge_width_scale = 0.1,
    save = "PAGA_results.png"
)
```
